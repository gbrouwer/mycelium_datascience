### Is Image Helper Function

```python
def isimage(item):
    
    imagefound = False
    for subitem in item:
        for subsubitem in subitem:
            if 'media/image' in subsubitem:
                imagefound = True
    
    return imagefound
```

### Is Caption Helper Function

```python
def iscaption(item):
    
    captionfound = False
    for subitem in item:
        for subsubitem in subitem:
            if 'Figure' in subsubitem:
                captionfound = True
    
    return captionfound
```
