### Read Stopwords

```python
stopwords = []
with open('../../data/meta/stopwords','r') as f:
    for line in f:
        stopwords.append(line.rstrip())
```

### Load

```python
data = pickle.load( open("../../data/breast_cancer_articles/pickle/processed.p", "rb"))
records = data['records']
```

### Extract Authors

```python
author_data = []
for record in records:
    title,abstract,authors = record
    author_data.append(authors)
```

### Create Author Graph

```python
graph = {}
for authors in tqdm(author_data):
    for author1 in authors:
        for author2 in authors:
            if author1 != author2:
                if author1 not in graph:
                    graph[author1] = [author2]
                else:
                    coauthors = graph[author1]
                    coauthors.append(author2)
                    graph[author1] = list(set(coauthors))
```
