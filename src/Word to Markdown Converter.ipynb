### Import

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from docx2python import docx2python
```

### Is Image Helper Function

```python
def isimage(item):
    
    imagefound = False
    for subitem in item:
        for subsubitem in subitem:
            if 'media/image' in subsubitem:
                imagefound = True
    
    return imagefound
```

### Is Caption Helper Function

```python
def iscaption(item):
    
    captionfound = False
    for subitem in item:
        for subsubitem in subitem:
            if 'Figure' in subsubitem:
                captionfound = True
    
    return captionfound
```

### Open Word File

```python
doc_result = docx2python('../word/001 - Foreword.docx')
```

### Process

```python
imagecounter = 1
chapterno = 1
for i in range(len(doc_result.body)):
    paragraph = doc_result.body[i]

    #First Paragraph
    if (i == 0):
        paragraph = paragraph[0][0]
        
        #Title
        print('#h1 ' + paragraph[0])
        
        #Subtitle
        print('#h2 ' + paragraph[1])
        print('')

        #Paragraph body
        for item in paragraph[2:]:
            
            #Test if it a subparagraph
            if item[0].isnumeric():
                print('#h3 ' + item)
            else:
                print('#pg ' + item)
                
    else:
        for item in paragraph:

            if isimage(item):
                imstr = '#im ../../assets/mycelium/figures/' + str(chapterno).zfill(3) + '/' + str(imagecounter).zfill(3) + '.png 50 256 '
                imagecounter = imagecounter + 1
            
            if iscaption(item):
                caption = item[1]
                print(imstr + caption[0])
                
            if (isimage(item) == False and iscaption(item) == False):

                #Test if it a subparagraph
                item = item[0][1]
                if item[0].isnumeric():
                    print('#h3 ' + item)
                else:
                    print('#pg ' + item)
                    print('')

    print('')
                

            
            
```
